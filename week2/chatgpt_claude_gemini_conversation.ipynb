```python

import os
import requests
from dotenv import load_dotenv
from IPython.display import Markdown, display
from anthropic import Anthropic
from google import genai
from openai import OpenAI


openai = OpenAI()
anthropic = Anthropic()
gemini = genai.Client()


gpt_model = "gpt-4.1-mini"
claude_model = "claude-haiku-4-5-20251001"
gemini_model = "gemini-2.5-flash"

gpt_system = "You are a Alex who is very argumentative; \
you disagree with anything in the conversation and you challenge everything, in a snarky way. you are in a conversation with Charlie and Alex."


def call_gpt(conversation):
    
    gpt_prompt = f"""
                    You are Alex, in conversation with Blake and Charlie.
                    The conversation so far is as follows:
                    {conversation}
                    Now with this, respond with what you would like to say next, as Alex.
                    """
    
    messages = [{"role": "system", "content": gpt_system}]
    messages.append({"role": "user", "content": gpt_prompt})
    response = openai.chat.completions.create(model=gpt_model, messages=messages)
    return response.choices[0].message.content

claude_system = "You are a very polite, courteous chatbot named Charlie. You try to agree with \
everything the other person says, or find common ground. If the other person is argumentative, \
you try to calm them down and keep chatting. You are in a conversation with Blake and Alex."


def call_claude(conversation):
    
    claude_prompt = f"""
                    You are Charlie, in conversation with Alex and Blake.
                    The conversation so far is as follows:
                    {conversation}
                    Now with this, respond with what you would like to say next, as Charlie.
                    """
    messages= [{"role": "user", "content": claude_prompt}]
    response = anthropic.messages.create(
        model=claude_model,
        system=claude_system,
        max_tokens=1024,
        messages=messages
    )
    return response.content[0].text

gemini_system = "You are a chatbot named Blake who is very argumentative; \
you disagree with anything in the conversation and you challenge everything, in a snarky way. You are in a conversation with Alex and Charlie."


def call_gemini(conversation):
    
    gemini_prompt = f"""
                        You are Blake, in conversation with Alex and Charlie.
                        The conversation so far is as follows:
                        {conversation}
                        Now with this, respond with what you would like to say next, as Blake.
                        """
    
    response = gemini.models.generate_content(
            model=gemini_model,
            contents=gemini_prompt,
            config=genai.types.GenerateContentConfig(
                system_instruction=gemini_system
            )
        )
    return response.text

conversation = [
                "Alex: Hi there", 
                "Charlie: Hi", 
                "Blake: Hello"
                ]

display(Markdown(f"\n{conversation[0]}\n"))
display(Markdown(f"\n{conversation[1]}\n"))
display(Markdown(f"\n{conversation[2]}\n"))

for i in range(5):
    gpt_next = call_gpt(conversation)
    conversation.append(f"Alex: {gpt_next}")
    display(Markdown(f"### Alex:\n{gpt_next}\n"))
    
    claude_next = call_claude(conversation)
    conversation.append(f"Charlie: {claude_next}")
    display(Markdown(f"### Charlie:\n{claude_next}\n"))
    
    gemini_next = call_gemini(conversation)
    conversation.append(f"Blake: {gemini_next}")
    display(Markdown(f"### Blake:\n{gemini_next}\n"))
```
